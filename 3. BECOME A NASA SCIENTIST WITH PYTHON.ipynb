```python
import glob #PACKAGE FOR READING FILE SYSTEM
from skimage.io import imread #SCIKIT IMAGE LIBRARY FOR LOADING IMAGES
from skimage.feature import blob_log #ALGORITHAM FOR IMAGEPIXELVALUE DISTRIBUTION
import matplotlib.pyplot as plt #PLOTTING
```

```python
star_file= glob.glob("PROJECT IMG 1.jpg")[0] #READING IMAGE FROM A PARTICULAR LOCATION
img= imread(star_file,as_gray=True)
```

```python
plt.rcParams['figure.figsize']=[8,12]
plt.imshow(img,cmap="gray")
```

```python
blobs_log=blob_log(img,max_sigma=30,num_sigma=10,threshold=0.1)
numrows= len(blobs_log)
```
